````python
import nltk
from nltk.translate.bleu_score import sentence_bleu

def compute_bleu(candidate, references):
    # 将候选翻译和参考翻译转换为标记化的列表
    candidate_tokens = candidate.split()
    reference_tokens = [ref.split() for ref in references]

    print(candidate_tokens)
    print(reference_tokens)
    
    # 使用 NLTK 中的 sentence_bleu 函数计算 BLEU 分数
    # weights 参数用于指定 BLEU 中不同 n-gram 的权重，默认为 (0.25, 0.25, 0.25, 0.25)
    # 使用 (1, 0, 0, 0) 权重表示只考虑 1-gram 的重叠
    bleu_score = sentence_bleu(reference_tokens, candidate_tokens, weights=(0.25, 0.25, 0.25, 0.25))
    
    return bleu_score

# 示例调用
candidate_translation = "\n private RequestService requestService ;\n  @ Override\n"
reference_translations = ["    @ Autowired\n    private RequestService requestService ;\n\n    @ Override\n"]
bleu_score = compute_bleu(candidate_translation, reference_translations)
print("BLEU-4 Score:", bleu_score)
````

````python
"""收集程序分析行号数据集"""

"""本质KMP算法"""
def findFirstIndex(haystack, needle):
    if needle == "":
        return 0

    index = haystack.find(needle)
    return index if index != -1 else -1

def read_lines_as_list(file_path, line_numbers):
    result = []
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
        lines = file.readlines()
        for line_number in line_numbers:
            if 1 <= line_number <= len(lines):
                result.append(lines[line_number - 1])
            else:
                result.append(line_number)

    return result

# 读取文件内容
def read_file_content(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except Exception as e:
        print(f"Error reading file: {e}")
    return None

# 统计行数
def count_newlines(input_string):
    newline_count = input_string.count('\n')
    return newline_count

# 生成所有行列表list
def generate_range(startline, endline):
    return list(range(startline, endline + 1))

def get_changeList(aob_path, aob_content):
    # 示例用法
    haystack = read_file_content(aob_path)
    needle = aob_content

    first_index = findFirstIndex(haystack, needle)
    # print("First occurrence of needle in haystack:", first_index)

    if first_index == -1:
        startline = -1
        endline = -1
    else:
        startline = count_newlines(haystack[:first_index]) + 1
        endline= startline + count_newlines(needle) - 1

    # 这里是左闭右闭！
    # print(f"第{startline}行 -> 第{endline}行")
    # print(generate_range(startline, endline))
    return generate_range(startline, endline)

# get_changeList(
#     "G:/now/2024merge/MergeBERT_Data/ChatGPTResearch/exampleData/acceptA/100004_a.java",
#     "        Platform.runLater(() -> {\n            //if there is an existing prompt or progressdialog,...\n            if (promptDialogManager.bringCurrentDialogToFront()) {\n                //... just show that\n            } else {\n\n                if ( //confirm timeline during ingest\n                        IngestManager.getInstance().isIngestRunning()\n                        && promptDialogManager.confirmDuringIngest() == false) {\n                    return;  //if they cancel, do nothing.\n                }\n"
# )

# file_path = 'G:/now/2024merge/Merge50Repo_Data/index_conflict_files/xtext/89_a.java'
# result_list = read_lines_as_list(file_path,  [74, 75])
# print(result_list)
````

````python
"""收集50Repo中100示例的 程序切片上下文"""
import json
from joern.mychange import get_program_dependence

def slicing_50repo_100(input_jsonPath, output_jsonPath, joern_path):

    file_path = "G:/now/2024merge/Merge50Repo_Data"
    with open(input_jsonPath, 'r', encoding='utf-8') as input_file:
        json_data = json.load(input_file)
    
    for i in range(len(json_data)):

        item = json_data[i]
        a_Path = file_path + item.get("file_name")[1:] + item.get("json_name")[:-13] + "a.java"
        o_Path = file_path + item.get("file_name")[1:] + item.get("json_name")[:-13] + "base.java"
        b_Path = file_path + item.get("file_name")[1:] + item.get("json_name")[:-13] + "b.java"
        print(a_Path)

        aob_lines = {}
        aob_lines["a_lines"] = get_changeList(a_Path, item.get("a_contents"))
        aob_lines["o_lines"] = get_changeList(o_Path, item.get("o_contents"))
        aob_lines["b_lines"] = get_changeList(b_Path, item.get("b_contents"))
        item["aob_lines"] = aob_lines

        print(f"""aob对应的行号：{aob_lines}""")
       
        item["key_information"] = []

        depth = 1 # depth 统一为1层
        aob_slicing = {}

        aob_slicing["a_slicing"] = get_program_dependence(
            joern_path + (str)(item.get("id")) + "_a.java.nodes.json", 
            joern_path + (str)(item.get("id")) + "_a.java.edges.json", 
            aob_lines["a_lines"], depth)[0]
        
        aob_slicing["o_slicing"] = get_program_dependence(
            joern_path + (str)(item.get("id")) + "_o.java.nodes.json", 
            joern_path + (str)(item.get("id")) + "_o.java.edges.json", 
            aob_lines["o_lines"], depth)[0]
        
        aob_slicing["b_slicing"] = get_program_dependence(
            joern_path + (str)(item.get("id")) + "_b.java.nodes.json", 
            joern_path+ (str)(item.get("id")) + "_b.java.edges.json", 
            aob_lines["b_lines"], depth)[0]

        item["aob_slicing"] = aob_slicing
        print(f"""第{i + 1}个切片结果为：{item["aob_slicing"]}""")

        aob_slicingContent = {}
        aob_slicingContent["a_slicingContext"] = read_lines_as_list(a_Path, aob_slicing["a_slicing"])
        aob_slicingContent["o_slicingContext"] = read_lines_as_list(o_Path, aob_slicing["o_slicing"])
        aob_slicingContent["b_slicingContext"] = read_lines_as_list(b_Path, aob_slicing["b_slicing"])
        # print(aob_slicingContent)
        item["key_context"] = aob_slicingContent

    with open(output_jsonPath, 'w', encoding='utf-8') as output_file:
                json.dump(json_data, output_file, indent=2)

slicing_50repo_100(input_jsonPath = "G:/now/2024merge/mergeMinePython/json/50Repo/50_100NC.json", 
                  output_jsonPath = "G:/now/2024merge/mergeMinePython/json/50Repo/50_100NC_slicing.json",
                  joern_path="G:/now/2024merge/mergeMinePython/joern/outputJson/50_100NC/") 
````

````python
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

def download_or_load_model(local_model_path):
    try:
        # 尝试从本地加载模型
        tokenizer = AutoTokenizer.from_pretrained(local_model_path)
        model = AutoModel.from_pretrained(local_model_path).to("cuda")
        print("使用本地模型")
    except:
        # 本地不存在则从网络下载
        tokenizer = AutoTokenizer.from_pretrained("microsoft/codebert-base")
        model = AutoModel.from_pretrained("microsoft/codebert-base")
        # 保存到本地
        tokenizer.save_pretrained(local_model_path)
        model.save_pretrained(local_model_path)
        print("从网络下载并保存本地模型")

    return tokenizer, model

def get_embedding(code, tokenizer, model):
    # 将代码转换为嵌入向量
    tokens = tokenizer(code, return_tensors="pt", padding=True, truncation=True).to("cuda")
    with torch.no_grad():
        output = model(**tokens)
    embeddings = output.last_hidden_state.mean(dim=1).to("cuda")  # 使用平均池化获取整个代码的嵌入
    return embeddings

def calculate_cosine_similarity(embedding1, embedding2):
    # 将张量移动到 CPU
    embedding1 = embedding1.cpu().numpy()
    embedding2 = embedding2.cpu().numpy()
    
    # 计算余弦相似性
    similarity = cosine_similarity(embedding1, embedding2)
    return similarity[0][0]
````

````python
 # 找到merged 文件 开始寻找代码空间距离最相似的代码行
def get_highest_score_codeLine(cur_mergedPath, key_info):
    #  errors='ignore' 在解码时忽略错误的字节，这样可以避免抛出异常，但可能导致一些字符丢失
    with open(cur_mergedPath, 'r', encoding='utf-8', errors='ignore') as file:
        lines = file.readlines()

    max_score = 0
    highest_line = ""
    key_info_embedding = get_embedding(key_info.strip(), tokenizer, model)
    
    skip = False
    for line in lines:
        if line.startswith("<<<<<<<"):
            skip = True
        elif line.startswith(">>>>>>>"):
            skip = False
        elif not skip:
            processed_line = line.strip()
            embedding = get_embedding(processed_line, tokenizer, model)
            cur_score = calculate_cosine_similarity(embedding, key_info_embedding)
            if cur_score > max_score:
                max_score = cur_score
                highest_line = processed_line

    # Print or use the Highest score line
    print(f"Highest {max_score} score Line:{highest_line}")
    return highest_line
````

````python
import re
def pre_conflict_suff(input_string):
    # 定义正则表达式模式
    pattern = re.compile(r'(?P<pre>.*?)<<<<<<<(?P<conflict>.*?)>>>>>>>.*?(?P<suff>.*)', re.DOTALL)

    # 使用正则表达式匹配字符串
    match = pattern.match(input_string)

    # 提取匹配的部分
    pre = match.group('pre')
    # conflict = match.group('conflict')
    suff = match.group('suff')

    # 输出结果
    return pre, suff
    # print("pre:", pre)
    # print("suff:", suff)
````

#### 使用正则表达式替换每一个冲突块

````python
import re
import json

def replace_code_blocks_with_json(input_str, json_data, choice):
    # 使用正则表达式匹配以<<<<<<<开始并以>>>>>>>结束的代码块
    pattern = re.compile(r"<<<<<<<\s*.*?\s*=======\s*(.*?)\s*>>>>>>>", re.DOTALL)
    
    # 将JSON数据解析为Python对象
    json_obj = json.loads(json_data)
    
    # 记录替换的代码块数
    block_count = 0
    
    # 替换每个匹配的代码块
    def replace_block(match):
        nonlocal block_count
        block_count += 1
        replacement = str(json_obj[block_count - 1].get(choice, ""))
        return replacement

    output_str = pattern.sub(replace_block, input_str)
    
    return output_str

````

````python
import difflib

"""行对齐有错误！"""
def align_and_label_lines(before_code, after_code):
    differ = difflib.Differ()
    diff = list(differ.compare(before_code.splitlines(), after_code.splitlines()))

    aligned_pairs = []
    li = 1

    for line in diff:
        code_change_type = line[0]
        code_line = line[2:]

        if code_change_type == ' ':
            # No change
            aligned_pairs.append((0, 0, code_line))
            li += 1
        elif code_change_type == '-':
            # Line deletion
            aligned_pairs.append((0, li, code_line))
            li += 1
        elif code_change_type == '+':
            # Line addition
            aligned_pairs.append((li, 0, code_line))
            li += 1
        elif code_change_type == '?':
            print(code_line)
            # Line replacement
            aligned_pairs.append((li, li, code_line))
            li += 1

    return aligned_pairs

# Example usage:
before_change_code = """if (cursor != null) {
    cursor.moveToFirst(); 
    int idx = cursor.getColumnIndex(); 
    result = cursor.getString(idx);}
"""

after_change_code = """if (cursor != null && cursor.moveToFirst()) {
    int idx = cursor.getColumnIndex();
    if (idx != -1) 
    result = cursor.getString(idx);}
"""

aligned_pairs = align_and_label_lines(before_change_code, after_change_code)

for pair in aligned_pairs:
    print(f"Before: {pair[0]}, After: {pair[1]}, Code: {pair[2]}")
````

#### token粒度对齐

````python
import difflib

"""此token对齐 应该没问题"""
def align_and_mark_changes(Ta, Tb):
    # Tokenize input strings
    tokens_a = Ta.split()
    tokens_b = Tb.split()

    # Initialize change flags
    ma = [0] * len(tokens_a)
    mb = [0] * len(tokens_b)

    # Use Ratcliff/Obershelp algorithm for token alignment
    matcher = difflib.SequenceMatcher(None, tokens_a, tokens_b)

    for tag, i1, i2, j1, j2 in matcher.get_opcodes():
        if tag == 'replace':
            # Tokens are different
            for i in range(i1, i2):
                ma[i] = 1
            for j in range(j1, j2):
                mb[j] = 1
        elif tag == 'delete':
            # Token present in Ta but not in Tb (deleted)
            for i in range(i1, i2):
                ma[i] = 1
        elif tag == 'insert':
            # Token present in Tb but not in Ta (added)
            for j in range(j1, j2):
                mb[j] = 1

    return ma, mb

# Example usage
Ta = "This is a very is easy  sample text for token alignment."
Tb = "This is a modified sample text for token alignment."
# Ta = "We align T b and T a token by token using Python difflib to identify the changed tokens."
# Tb = "We align T a and T b token by token using Python difflib to identify the modified tokens."


ma, mb = align_and_mark_changes(Ta, Tb)

print("Token alignment:")
print(f"  Ta: {Ta}")
print(f"  Tb: {Tb}")
print(f"  ma: {ma}")
print(f"  mb: {mb}")
````

````python
"""重写json 文件 从而重置其属性顺序"""
import json
def reset_order(input_jsonPath, output_jsonPath):
    # read from file
    with open(input_jsonPath, 'r', encoding='utf-8') as file:
        json_data = json.load(file)

    new_json_data = []
    for item in json_data:
        new_item = {
            "id": item["id"],
            "json_name": item["json_name"],
            # "repo": item["repo"],
            "file_name": item["file_name"],
            "id_inFile": item["id_inFile"],
            "a_contents": item["a_contents"],
            "o_contents": item["o_contents"],
            "b_contents": item["b_contents"],
            "res_label": item["res_label"], 
            "can_token_level": item["can_token_level"],
            "can_merge_succeed": item["can_merge_succeed"],
            "token_level_result": item["token_level_result"],
            "match_rate": item["match_rate"],
            "key_information": item["key_information"],
            "key_context": item["key_context"],
            "res_region": item["res_region"],
            "line_noContext_answer": item["line_noContext_answer"],
            "line_withContext_answer": item["line_withContext_answer"],
            "token_noContext_answer": item["token_noContext_answer"],
            "token_withContext_answer": item["token_withContext_answer"],
            "line_no_maxMatch": item["line_no_maxMatch"],
            "line_with_maxMatch": item["line_with_maxMatch"],
            "token_no_maxMatch": item["token_no_maxMatch"],
            "token_with_maxMatch": item["token_with_maxMatch"]
        }
        new_json_data.append(new_item)

    # write in file
    with open(output_jsonPath, 'w', encoding='utf-8') as file:
        json.dump(new_json_data, file, indent=2)

"""调用使用重写json顺序方法"""
reset_order(
    input_jsonPath = "G:/now/2024merge/mergeMinePython/output/50Repo_Output/50_100v2_COT.json", 
    output_jsonPath ="G:/now/2024merge/mergeMinePython/output/50Repo_Output/50_100v2_COT.json")
````
